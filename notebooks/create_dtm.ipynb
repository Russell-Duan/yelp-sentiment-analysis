```python
X_train = data_train['review'].tolist()
y_train = data_train['sentiment'].tolist()
```

```python
X_test = data_test['review'].tolist()
y_test = data_test['sentiment'].tolist()
```

# Vectorize

- CountVectorizer
- Unigram + Bigram
- Remove non-words/numbers
- Remove stopwords
- Lemmatization

```python
import nltk
```

```python
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')
```

```python
vect = features.YelpSentCountVectorizer(ngram_range=(1,2),
                                        remove_nonwords=True,
                                        remove_stopwords=True,
                                        stem=False,
                                        lemmatize=True)
```

```python
%time cv = vect.fit(X_train)
```

```python
with open('drive/My Drive/yelp_sentiment_analysis/pickles/cv.pickle', 'wb') as f:
  pickle.dump(cv, f)
```

```python
X_train_dtm = vect.transform(X_train)
X_test_dtm = vect.transform(X_test)
```

```python
with open('drive/My Drive/yelp_sentiment_analysis/pickles/X_train_dtm.pickle', 'wb') as f:
  pickle.dump(X_train_dtm, f)

with open('drive/My Drive/yelp_sentiment_analysis/pickles/X_test_dtm.pickle', 'wb') as f:
  pickle.dump(X_test_dtm, f)
```
